```python
# =============================================================================
# CELL 3: Data Preprocessing and Feature Engineering (FIXED)
# =============================================================================

def preprocess_data(df):
    """
    Preprocess the dataset for dynamic pricing
    """
    # Create a copy to avoid modifying original
    data = df.copy()

    # Convert date and time columns to datetime (CORRECTED FORMAT)
    # Original format: "13-10-2016" (DD-MM-YYYY)
    data['DateTime'] = pd.to_datetime(
        data['LastUpdatedDate'] + ' ' + data['LastUpdatedTime'],
        format='%d-%m-%Y %H:%M:%S'  # Corrected format specifier
    )

    # Calculate occupancy rate
    data['OccupancyRate'] = data['Occupancy'] / data['Capacity']

    # Encode vehicle type with weights (pricing factor)
    vehicle_weights = {
        'Car': 1.0,
        'Bike': 0.5,
        'Truck': 1.5
    }
    data['VehicleTypeWeight'] = data['VehicleType'].map(vehicle_weights)

    # Encode traffic conditions
    traffic_levels = {
        'Low': 0.5,
        'Medium': 1.0,
        'High': 1.5
    }
    data['TrafficLevel'] = data['TrafficConditionNearby'].map(traffic_levels)

    # Sort by DateTime for time series processing
    data = data.sort_values(['SystemCodeNumber', 'DateTime'])

    # Initialize base price
    data['BasePrice'] = 10.0

    return data

# Preprocess the data
processed_df = preprocess_data(df)
print("✅ Data preprocessing completed!")
print(f"Processed data shape: {processed_df.shape}")
print(f"Date range: {processed_df['DateTime'].min()} to {processed_df['DateTime'].max()}")
```

```python
# CELL 4: Utility Functions
# =============================================================================

def haversine_distance(lat1, lon1, lat2, lon2):
    """
    Calculate the great circle distance between two points
    on the earth (specified in decimal degrees)
    """
    # Convert decimal degrees to radians
    lat1, lon1, lat2, lon2 = map(math.radians, [lat1, lon1, lat2, lon2])

    # Haversine formula
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = math.sin(dlat/2)**2 + math.cos(lat1) * math.cos(lat2) * math.sin(dlon/2)**2
    c = 2 * math.asin(math.sqrt(a))

    # Radius of earth in kilometers
    r = 6371
    return c * r

def get_nearby_lots(df, target_lat, target_lon, radius_km=2.0):
    """
    Find parking lots within a specified radius
    """
    nearby_lots = []
    for _, row in df.iterrows():
        distance = haversine_distance(target_lat, target_lon, row['Latitude'], row['Longitude'])
        if distance <= radius_km:
            nearby_lots.append({
                'SystemCodeNumber': row['SystemCodeNumber'],
                'Distance': distance,
                'Latitude': row['Latitude'],
                'Longitude': row['Longitude']
            })
    return nearby_lots

print("✅ Utility functions defined!")
```

```python
# CELL 5: Model 1 - Baseline Linear Model
# =============================================================================

class Model1_LinearPricing:
    def __init__(self, alpha=5.0, base_price=10.0):
        """
        Model 1: Simple linear pricing based on occupancy

        Parameters:
        - alpha: Price sensitivity to occupancy rate
        - base_price: Starting price
        """
        self.alpha = alpha
        self.base_price = base_price
        self.price_history = {}

    def calculate_price(self, parking_lot_id, occupancy_rate, previous_price=None):
        """
        Calculate new price based on occupancy rate
        """
        if previous_price is None:
            previous_price = self.base_price

        # Linear price adjustment
        price_adjustment = self.alpha * occupancy_rate
        new_price = previous_price + price_adjustment

        # Ensure price bounds (0.5x to 2x base price)
        min_price = self.base_price * 0.5
        max_price = self.base_price * 2.0
        new_price = max(min_price, min(max_price, new_price))

        # Store in history
        self.price_history[parking_lot_id] = new_price

        return new_price

# Test Model 1
model1 = Model1_LinearPricing(alpha=5.0)

# Apply Model 1 to a sample parking lot
sample_lot = processed_df[processed_df['SystemCodeNumber'] == processed_df['SystemCodeNumber'].iloc[0]]
sample_lot = sample_lot.sort_values('DateTime')

model1_prices = []
for _, row in sample_lot.iterrows():
    prev_price = model1_prices[-1] if model1_prices else None
    price = model1.calculate_price(
        row['SystemCodeNumber'],
        row['OccupancyRate'],
        prev_price
    )
    model1_prices.append(price)

print("✅ Model 1 implemented and tested!")
print(f"Sample prices for first lot: {model1_prices[:10]}")
```

```python
# CELL 5: Model 1 - Baseline Linear Model
# =============================================================================

class Model1_LinearPricing:
    def __init__(self, alpha=5.0, base_price=10.0):
        """
        Model 1: Simple linear pricing based on occupancy

        Parameters:
        - alpha: Price sensitivity to occupancy rate
        - base_price: Starting price
        """
        self.alpha = alpha
        self.base_price = base_price
        self.price_history = {}

    def calculate_price(self, parking_lot_id, occupancy_rate, previous_price=None):
        """
        Calculate new price based on occupancy rate
        """
        if previous_price is None:
            previous_price = self.base_price

        # Linear price adjustment
        price_adjustment = self.alpha * occupancy_rate
        new_price = previous_price + price_adjustment

        # Ensure price bounds (0.5x to 2x base price)
        min_price = self.base_price * 0.5
        max_price = self.base_price * 2.0
        new_price = max(min_price, min(max_price, new_price))

        # Store in history
        self.price_history[parking_lot_id] = new_price

        return new_price

# Test Model 1
model1 = Model1_LinearPricing(alpha=5.0)

# Apply Model 1 to a sample parking lot
sample_lot = processed_df[processed_df['SystemCodeNumber'] == processed_df['SystemCodeNumber'].iloc[0]]
sample_lot = sample_lot.sort_values('DateTime')

model1_prices = []
for _, row in sample_lot.iterrows():
    prev_price = model1_prices[-1] if model1_prices else None
    price = model1.calculate_price(
        row['SystemCodeNumber'],
        row['OccupancyRate'],
        prev_price
    )
    model1_prices.append(price)

print("✅ Model 1 implemented and tested!")
print(f"Sample prices for first lot: {model1_prices[:10]}")
```

```python
# CELL 6: Model 2 - Demand-Based Pricing
# =============================================================================

class Model2_DemandBasedPricing:
    def __init__(self, alpha=2.0, beta=1.5, gamma=1.0, delta=3.0, epsilon=2.0, lambda_price=0.3, base_price=10.0):
        """
        Model 2: Demand-based pricing

        Parameters:
        - alpha: Occupancy rate coefficient
        - beta: Queue length coefficient
        - gamma: Traffic level coefficient (negative impact)
        - delta: Special day coefficient
        - epsilon: Vehicle type weight coefficient
        - lambda_price: Price adjustment factor
        - base_price: Starting price
        """
        self.alpha = alpha
        self.beta = beta
        self.gamma = gamma
        self.delta = delta
        self.epsilon = epsilon
        self.lambda_price = lambda_price
        self.base_price = base_price

    def calculate_demand(self, occupancy_rate, queue_length, traffic_level, is_special_day, vehicle_type_weight):
        """
        Calculate demand score based on multiple factors
        """
        raw_demand = (
            self.alpha * occupancy_rate +
            self.beta * queue_length -
            self.gamma * traffic_level +
            self.delta * is_special_day +
            self.epsilon * vehicle_type_weight
        )

        # Normalize demand (sigmoid-like function)
        normalized_demand = np.tanh(raw_demand / 5.0)

        return normalized_demand

    def calculate_price(self, occupancy_rate, queue_length, traffic_level, is_special_day, vehicle_type_weight):
        """
        Calculate price based on demand
        """
        demand = self.calculate_demand(occupancy_rate, queue_length, traffic_level, is_special_day, vehicle_type_weight)

        # Price adjustment
        price = self.base_price * (1 + self.lambda_price * demand)

        # Ensure price bounds
        min_price = self.base_price * 0.5
        max_price = self.base_price * 2.0
        price = max(min_price, min(max_price, price))

        return price, demand

# Test Model 2
model2 = Model2_DemandBasedPricing()

# Apply Model 2 to sample data
model2_prices = []
model2_demands = []

for _, row in sample_lot.iterrows():
    price, demand = model2.calculate_price(
        row['OccupancyRate'],
        row['QueueLength'],
        row['TrafficLevel'],
        row['IsSpecialDay'],
        row['VehicleTypeWeight']
    )
    model2_prices.append(price)
    model2_demands.append(demand)

print("✅ Model 2 implemented and tested!")
print(f"Sample prices for first lot: {model2_prices[:10]}")
print(f"Sample demands for first lot: {model2_demands[:10]}")
```

```python
# CELL 7: Model 3 - Competitive Pricing
# =============================================================================

class Model3_CompetitivePricing:
    def __init__(self, model2_instance, base_price=10.0):
        """
        Model 3: Competitive pricing with location intelligence

        Parameters:
        - model2_instance: Instance of Model2 for base pricing
        - base_price: Starting price
        """
        self.model2 = model2_instance
        self.base_price = base_price
        self.lot_locations = {}
        self.current_prices = {}

    def update_lot_location(self, lot_id, latitude, longitude):
        """
        Update parking lot location information
        """
        self.lot_locations[lot_id] = (latitude, longitude)

    def get_competitor_prices(self, target_lot_id, radius_km=2.0):
        """
        Get prices of nearby competitor lots
        """
        if target_lot_id not in self.lot_locations:
            return []

        target_lat, target_lon = self.lot_locations[target_lot_id]
        competitor_prices = []

        for lot_id, (lat, lon) in self.lot_locations.items():
            if lot_id != target_lot_id:
                distance = haversine_distance(target_lat, target_lon, lat, lon)
                if distance <= radius_km and lot_id in self.current_prices:
                    competitor_prices.append({
                        'lot_id': lot_id,
                        'price': self.current_prices[lot_id],
                        'distance': distance
                    })

        return competitor_prices

    def calculate_competitive_price(self, lot_id, occupancy_rate, queue_length, traffic_level, is_special_day, vehicle_type_weight):
        """
        Calculate price considering competition
        """
        # Get base price from Model 2
        base_price, demand = self.model2.calculate_price(
            occupancy_rate, queue_length, traffic_level, is_special_day, vehicle_type_weight
        )

        # Get competitor prices
        competitors = self.get_competitor_prices(lot_id)

        if not competitors:
            # No competitors nearby, use base price
            final_price = base_price
        else:
            # Calculate average competitor price
            avg_competitor_price = np.mean([comp['price'] for comp in competitors])

            # Competitive adjustment logic
            if occupancy_rate > 0.9:  # High occupancy
                if avg_competitor_price < base_price * 0.9:
                    # Competitors much cheaper, reduce price slightly
                    final_price = base_price * 0.95
                elif avg_competitor_price > base_price * 1.1:
                    # Competitors more expensive, increase price
                    final_price = base_price * 1.05
                else:
                    final_price = base_price
            else:  # Normal/low occupancy
                if avg_competitor_price > base_price * 1.1:
                    # Competitors expensive, increase price
                    final_price = base_price * 1.1
                elif avg_competitor_price < base_price * 0.9:
                    # Competitors cheap, reduce price
                    final_price = base_price * 0.9
                else:
                    final_price = base_price

        # Ensure price bounds
        min_price = self.base_price * 0.5
        max_price = self.base_price * 2.0
        final_price = max(min_price, min(max_price, final_price))

        # Update current price
        self.current_prices[lot_id] = final_price

        return final_price, demand, len(competitors)

# Test Model 3
model3 = Model3_CompetitivePricing(model2)

# Initialize lot locations
unique_lots = processed_df[['SystemCodeNumber', 'Latitude', 'Longitude']].drop_duplicates()
for _, row in unique_lots.iterrows():
    model3.update_lot_location(row['SystemCodeNumber'], row['Latitude'], row['Longitude'])

print("✅ Model 3 implemented and tested!")
print(f"Initialized {len(unique_lots)} parking lot locations")
```

```python
# CELL 8: Real-time Simulation with Pathway
# =============================================================================

# Since Pathway requires specific setup, let's create a simplified simulation
class RealTimeSimulation:
    def __init__(self, data, models):
        """
        Real-time simulation of dynamic pricing
        """
        self.data = data
        self.models = models
        self.results = {
            'timestamps': [],
            'lot_ids': [],
            'model1_prices': [],
            'model2_prices': [],
            'model3_prices': [],
            'occupancy_rates': [],
            'demands': [],
            'competitor_counts': []
        }

    def run_simulation(self, lot_id=None, max_steps=100):
        """
        Run the simulation for a specific lot or all lots
        """
        if lot_id:
            data_subset = self.data[self.data['SystemCodeNumber'] == lot_id].head(max_steps)
        else:
            data_subset = self.data.head(max_steps)

        model1, model2, model3 = self.models

        for idx, (_, row) in enumerate(data_subset.iterrows()):
            # Model 1 price
            prev_price = self.results['model1_prices'][-1] if self.results['model1_prices'] else None
            price1 = model1.calculate_price(row['SystemCodeNumber'], row['OccupancyRate'], prev_price)

            # Model 2 price
            price2, demand = model2.calculate_price(
                row['OccupancyRate'], row['QueueLength'], row['TrafficLevel'],
                row['IsSpecialDay'], row['VehicleTypeWeight']
            )

            # Model 3 price
            price3, demand3, comp_count = model3.calculate_competitive_price(
                row['SystemCodeNumber'], row['OccupancyRate'], row['QueueLength'],
                row['TrafficLevel'], row['IsSpecialDay'], row['VehicleTypeWeight']
            )

            # Store results
            self.results['timestamps'].append(row['DateTime'])
            self.results['lot_ids'].append(row['SystemCodeNumber'])
            self.results['model1_prices'].append(price1)
            self.results['model2_prices'].append(price2)
            self.results['model3_prices'].append(price3)
            self.results['occupancy_rates'].append(row['OccupancyRate'])
            self.results['demands'].append(demand)
            self.results['competitor_counts'].append(comp_count)

            # Simulate real-time delay
            if idx % 10 == 0:
                print(f"Processed {idx+1}/{len(data_subset)} records...")

        return self.results

# Create simulation instance
simulation = RealTimeSimulation(processed_df, [model1, model2, model3])

# Run simulation for first parking lot
first_lot_id = processed_df['SystemCodeNumber'].iloc[0]
print(f"Running simulation for parking lot: {first_lot_id}")
results = simulation.run_simulation(lot_id=first_lot_id, max_steps=50)

print("✅ Real-time simulation completed!")
```

```python
# CELL 9: Visualization with Bokeh
# =============================================================================

def create_pricing_visualization(results):
    """
    Create interactive Bokeh visualizations
    """
    # Convert timestamps to strings for Bokeh
    timestamps = [str(ts) for ts in results['timestamps']]

    # Create data source
    source = ColumnDataSource(data=dict(
        x=list(range(len(timestamps))),
        timestamps=timestamps,
        model1_prices=results['model1_prices'],
        model2_prices=results['model2_prices'],
        model3_prices=results['model3_prices'],
        occupancy_rates=results['occupancy_rates'],
        demands=results['demands']
    ))

    # Create main pricing plot
    pricing_plot = figure(
        title="Dynamic Pricing Models Comparison",
        x_axis_label="Time Step",
        y_axis_label="Price ($)",
        width=800,
        height=400,
        tools="pan,wheel_zoom,box_zoom,reset,save"
    )

    # Add lines for each model
    pricing_plot.line('x', 'model1_prices', source=source, legend_label="Model 1 (Linear)",
                     line_width=2, color='blue')
    pricing_plot.line('x', 'model2_prices', source=source, legend_label="Model 2 (Demand)",
                     line_width=2, color='red')
    pricing_plot.line('x', 'model3_prices', source=source, legend_label="Model 3 (Competitive)",
                     line_width=2, color='green')

    # Add hover tool
    hover = HoverTool(tooltips=[
        ("Time", "@timestamps"),
        ("Model 1 Price", "@model1_prices{$0.00}"),
        ("Model 2 Price", "@model2_prices{$0.00}"),
        ("Model 3 Price", "@model3_prices{$0.00}"),
        ("Occupancy Rate", "@occupancy_rates{0.00%}"),
        ("Demand", "@demands{0.00}")
    ])
    pricing_plot.add_tools(hover)

    # Create occupancy plot
    occupancy_plot = figure(
        title="Occupancy Rate Over Time",
        x_axis_label="Time Step",
        y_axis_label="Occupancy Rate",
        width=800,
        height=300,
        tools="pan,wheel_zoom,box_zoom,reset,save"
    )

    occupancy_plot.line('x', 'occupancy_rates', source=source, line_width=2, color='orange')

    # Create demand plot
    demand_plot = figure(
        title="Demand Score Over Time",
        x_axis_label="Time Step",
        y_axis_label="Demand Score",
        width=800,
        height=300,
        tools="pan,wheel_zoom,box_zoom,reset,save"
    )

    demand_plot.line('x', 'demands', source=source, line_width=2, color='purple')

    # Create layout
    layout = gridplot([[pricing_plot], [occupancy_plot], [demand_plot]])

    return layout

# Create and show visualizations
print("Creating visualizations...")
viz_layout = create_pricing_visualization(results)
show(viz_layout)

print("✅ Visualizations created!")
```

```python
# CELL 10: Analysis and Results
# =============================================================================

def analyze_results(results):
    """
    Analyze the pricing results
    """
    results_df = pd.DataFrame(results)

    print("=== PRICING ANALYSIS ===")
    print(f"Total time steps analyzed: {len(results_df)}")
    print(f"Parking lot: {results_df['lot_ids'].iloc[0]}")
    print()

    # Price statistics
    print("Price Statistics:")
    for model in ['model1_prices', 'model2_prices', 'model3_prices']:
        prices = results_df[model]
        print(f"{model}:")
        print(f"  Mean: ${prices.mean():.2f}")
        print(f"  Std:  ${prices.std():.2f}")
        print(f"  Min:  ${prices.min():.2f}")
        print(f"  Max:  ${prices.max():.2f}")
        print()

    # Correlation analysis
    print("Correlation with Occupancy Rate:")
    for model in ['model1_prices', 'model2_prices', 'model3_prices']:
        corr = results_df[model].corr(results_df['occupancy_rates'])
        print(f"{model}: {corr:.3f}")
    print()

    # Price volatility (standard deviation)
    print("Price Volatility (smoothness):")
    for model in ['model1_prices', 'model2_prices', 'model3_prices']:
        prices = results_df[model]
        volatility = prices.std() / prices.mean()
        print(f"{model}: {volatility:.3f}")
    print()

    # Demand statistics
    print("Demand Statistics:")
    demands = results_df['demands']
    print(f"Mean demand: {demands.mean():.3f}")
    print(f"Demand range: [{demands.min():.3f}, {demands.max():.3f}]")

    return results_df

# Analyze results
analysis_df = analyze_results(results)
```

```python
# CELL 14: Advanced Features and Extensions
# =============================================================================

class AdvancedPricingSystem:
    """
    Advanced pricing system with additional features
    """

    def __init__(self):
        self.model1 = Model1_LinearPricing(alpha=5.0)
        self.model2 = Model2_DemandBasedPricing()
        self.model3 = Model3_CompetitivePricing(self.model2)
        self.price_history = {}
        self.recommendations = []

    def suggest_rerouting(self, current_lot_id, occupancy_rate, threshold=0.9):
        """
        Suggest alternative parking lots when current lot is full
        """
        if occupancy_rate < threshold:
            return None

        # Get competitor information
        competitors = self.model3.get_competitor_prices(current_lot_id)

        if not competitors:
            return None

        # Find cheaper alternatives
        current_price = self.model3.current_prices.get(current_lot_id, 10.0)

        alternatives = []
        for comp in competitors:
            if comp['price'] < current_price:
                alternatives.append({
                    'lot_id': comp['lot_id'],
                    'price': comp['price'],
                    'distance': comp['distance'],
                    'savings': current_price - comp['price']
                })

        # Sort by savings (highest first)
        alternatives.sort(key=lambda x: x['savings'], reverse=True)

        return alternatives[:3]  # Return top 3 alternatives

    def dynamic_pricing_decision(self, lot_data):
        """
        Make comprehensive pricing decision
        """
        # Calculate all model prices
        price1 = self.model1.calculate_price(
            lot_data['lot_id'],
            lot_data['occupancy_rate']
        )

        price2, demand = self.model2.calculate_price(
            lot_data['occupancy_rate'],
            lot_data['queue_length'],
            lot_data['traffic_level'],
            lot_data['is_special_day'],
            lot_data['vehicle_type_weight']
        )

        price3, demand3, comp_count = self.model3.calculate_competitive_price(
            lot_data['lot_id'],
            lot_data['occupancy_rate'],
            lot_data['queue_length'],
            lot_data['traffic_level'],
            lot_data['is_special_day'],
            lot_data['vehicle_type_weight']
        )

        # Choose final price (can be weighted combination or business logic)
        if comp_count > 0:
            final_price = price3  # Use competitive pricing when competitors exist
        else:
            final_price = price2  # Use demand-based pricing otherwise

        # Check for rerouting recommendations
        alternatives = self.suggest_rerouting(
            lot_data['lot_id'],
            lot_data['occupancy_rate']
        )

        decision = {
            'lot_id': lot_data['lot_id'],
            'final_price': final_price,
            'model1_price': price1,
            'model2_price': price2,
            'model3_price': price3,
            'demand_score': demand,
            'competitor_count': comp_count,
            'alternatives': alternatives,
            'timestamp': datetime.now()
        }

        return decision

# Test advanced system
advanced_system = AdvancedPricingSystem()

# Initialize locations
for _, row in unique_lots_df.iterrows():
    advanced_system.model3.update_lot_location(row['SystemCodeNumber'], row['Latitude'], row['Longitude'])

# Test with sample data
sample_lot_data = {
    'lot_id': first_lot_id,
    'occupancy_rate': 0.95,
    'queue_length': 5,
    'traffic_level': 1.5,
    'is_special_day': 1,
    'vehicle_type_weight': 1.0
}

advanced_decision = advanced_system.dynamic_pricing_decision(sample_lot_data)

print("=== ADVANCED PRICING DECISION ===")
print(f"Lot ID: {advanced_decision['lot_id']}")
print(f"Final Price: ${advanced_decision['final_price']:.2f}")
print(f"Model Prices: M1=${advanced_decision['model1_price']:.2f}, M2=${advanced_decision['model2_price']:.2f}, M3=${advanced_decision['model3_price']:.2f}")
print(f"Demand Score: {advanced_decision['demand_score']:.3f}")
print(f"Competitors Found: {advanced_decision['competitor_count']}")

if advanced_decision['alternatives']:
    print("\nRerouting Recommendations:")
    for i, alt in enumerate(advanced_decision['alternatives'], 1):
        print(f"  {i}. Lot {alt['lot_id']}: ${alt['price']:.2f} (Save ${alt['savings']:.2f}, {alt['distance']:.2f}km away)")
else:
    print("\nNo rerouting recommendations available.")
```

```python
# CELL 15: Real-time Dashboard Simulation
# =============================================================================

from bokeh.models import Button, Div
from bokeh.layouts import column, row
from bokeh.io import curdoc
import threading
import time

class RealTimeDashboard:
    """
    Real-time dashboard for monitoring pricing
    """

    def __init__(self, pricing_system, data):
        self.pricing_system = pricing_system
        self.data = data
        self.current_index = 0
        self.dashboard_data = {
            'timestamps': [],
            'prices': [],
            'occupancy': [],
            'demand': [],
            'competitors': []
        }
        self.max_points = 50  # Keep last 50 data points

    def update_dashboard(self):
        """
        Update dashboard with new data point
        """
        if self.current_index >= len(self.data):
            return False

        # Get current data point
        current_row = self.data.iloc[self.current_index]

        # Prepare lot data
        lot_data = {
            'lot_id': current_row['SystemCodeNumber'],
            'occupancy_rate': current_row['OccupancyRate'],
            'queue_length': current_row['QueueLength'],
            'traffic_level': current_row['TrafficLevel'],
            'is_special_day': current_row['IsSpecialDay'],
            'vehicle_type_weight': current_row['VehicleTypeWeight']
        }

        # Get pricing decision
        decision = self.pricing_system.dynamic_pricing_decision(lot_data)

        # Update dashboard data
        self.dashboard_data['timestamps'].append(current_row['DateTime'])
        self.dashboard_data['prices'].append(decision['final_price'])
        self.dashboard_data['occupancy'].append(current_row['OccupancyRate'])
        self.dashboard_data['demand'].append(decision['demand_score'])
        self.dashboard_data['competitors'].append(decision['competitor_count'])

        # Keep only last max_points
        for key in self.dashboard_data:
            if len(self.dashboard_data[key]) > self.max_points:
                self.dashboard_data[key] = self.dashboard_data[key][-self.max_points:]

        self.current_index += 1
        return True

    def get_dashboard_html(self):
        """
        Generate HTML summary for dashboard
        """
        if not self.dashboard_data['prices']:
            return "<div>No data available</div>"

        current_price = self.dashboard_data['prices'][-1]
        current_occupancy = self.dashboard_data['occupancy'][-1]
        current_demand = self.dashboard_data['demand'][-1]
        current_competitors = self.dashboard_data['competitors'][-1]

        html = f"""
        <div style="padding: 20px; background-color: #f0f0f0; border-radius: 10px;">
            <h2>Real-Time Pricing Dashboard</h2>
            <div style="display: flex; justify-content: space-between;">
                <div style="text-align: center;">
                    <h3>Current Price</h3>
                    <p style="font-size: 24px; color: #2E86AB;">${current_price:.2f}</p>
                </div>
                <div style="text-align: center;">
                    <h3>Occupancy Rate</h3>
                    <p style="font-size: 24px; color: #A23B72;">{current_occupancy:.1%}</p>
                </div>
                <div style="text-align: center;">
                    <h3>Demand Score</h3>
                    <p style="font-size: 24px; color: #F18F01;">{current_demand:.3f}</p>
                </div>
                <div style="text-align: center;">
                    <h3>Competitors</h3>
                    <p style="font-size: 24px; color: #C73E1D;">{current_competitors}</p>
                </div>
            </div>
            <p style="margin-top: 20px; font-size: 12px; color: #666;">
                Last updated: {self.dashboard_data['timestamps'][-1]}
            </p>
        </div>
        """
        return html

# Create dashboard instance
dashboard_data = processed_df[processed_df['SystemCodeNumber'] == first_lot_id].head(100)
dashboard = RealTimeDashboard(advanced_system, dashboard_data)

# Simulate real-time updates
print("=== REAL-TIME DASHBOARD SIMULATION ===")
for i in range(10):
    if dashboard.update_dashboard():
        print(f"Update {i+1}: Price=${dashboard.dashboard_data['prices'][-1]:.2f}, "
              f"Occupancy={dashboard.dashboard_data['occupancy'][-1]:.1%}")
        time.sleep(0.5)  # Simulate real-time delay
    else:
        break

# Show final dashboard HTML
final_html = dashboard.get_dashboard_html()
display(HTML(final_html))
```
